```python
# generateTimesFile.sh makes lines of the form:
# time for run # 0 on test/simple.txt : 0.00680994987488
import pandas as pd
def timesFileToCSV(times_file_path):
    
    with open(times_file_path, 'r') as f:
        times_content = f.read()
    lines =times_content.split('\n')
    split_lines = [line.split(':')[1] for line in lines[:-1]]
    split_lines = [line.split(',') for line in split_lines]
    
    times_dict = [
        {"file":row[1],"time":float(row[2]),"path":times_file_path} 
            for row in split_lines]
    return pd.DataFrame(times_dict)
```

```python
d = timesFileToCSV("times_100_2000_1_10.txt")
d
```

```python
def get_all_experiments(paths):
    return pd.concat([timesFileToCSV(file_path) for file_path in paths], ignore_index=True)
```

```python
df = get_all_experiments(["times_1000_500000_1_10.txt",
                         "times_1000_500000_2_10.txt",
                         "times_1000_500000_4_10.txt",
                         "times_1000_500000_8_10.txt",
                         "times_1000_500000_16_10.txt",
                         "times_1000_500000_32_10.txt",
                         "times_1000_500000_64_10.txt",
                         "times_1000_500000_128_10.txt",])
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
plt.title("tiempo en funcion de experimento")
sns.lineplot(data=df, x="file", y="time", hue="path")
#plt.savefig("image_name")
```

```python
def add_exec_data(df):
    file_names = df['path']
    df['n'] = file_names.apply(lambda s: int(s.split("_")[1]))
    df['m'] = file_names.apply(lambda s: int(s.split("_")[2]))
    df['k'] = file_names.apply(lambda s: int(s.split("_")[3]))
    df['w'] = file_names.apply(lambda s: int(s.split("_")[4][:-4]))
    return df
```

```python
add_exec_data(df)
```

```python
plt.title("tiempo (segundos) en funcion de cantidad de threads, n=1000 m=500000")
sns.lineplot(data=df, x="k", y="time")
plt.savefig("time_plot_1000_500000.png")
```
